# 4. Measure Of Dispersion
	1. Variance
	2. Standard Deviation

```python
print(np.var(dataset)) # Uses Population Varience Formula where is ÷ by n
print(statistics.pvariance(dataset)) # Uses Population Varience Formula where is ÷ by n
print(statistics.variance(dataset)) # Uses Sample Varience Formula where is ÷ by n-1
```

### 1. How Do I Calculate Variance?
1. Calculate the mean of the data.
2. Find each data point's difference from the mean value.
3. Square each of these values.
4. Add up all of the squared values.
5. Divide this sum of squares by n – 1 (for a sample) or N (for the population).

```python
def variance(data):
    n=len(data);
    mean=sum(data)/n
    deviation=[(mean-i)**2 for i in data]
    return sum(deviation)/n-1
```

```python
variance(dataset)
```

### 2. Calculate Standard Deviation

```python
import math
math.sqrt(variance(dataset))
```

### 4. Check a Distribution is Normal or Not

```python
### If you want to check whether a feature is normal or LogNormal Distribution
### Q-Q Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['petal_width'])
```
